```python
sales_df = pd.read_csv('sales-data-sample.csv')
print(sales_df.head())
```

```python
sales_by_region = sales_df.groupby('Region')['Sales'].sum()
print(sales_by_region)
```

```python
best_selling_products = sales_df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False)
print(best_selling_products)
```

```python
sales_by_region.to_csv('sales_by_region.csv')
best_selling_products.to_csv('best_selling_products.csv')
print("Analysis results saved to CSV files.")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
data = yf.download(tickers, start="2024-01-01", end="2024-12-30", group_by='ticker')
print(data.head())
```

```python
for ticker in tickers:
    data[ticker, 'Daily Return'] = data[ticker, 'Close'].pct_change()
```

```python
plt.figure(figsize=(12, 6))

for i, ticker in enumerate(tickers):
    plt.subplot(3, 1, i+1)
    data[(ticker, 'Close')].plot(title=f"{ticker} - Closing Price")

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))

for i, ticker in enumerate(tickers):
    plt.subplot(3, 1, i+1)
    data[(ticker, 'Daily Return')].plot(title=f"{ticker} - Daily Returns", color='orange')

plt.tight_layout()
plt.show()
```

```python
data.to_csv('bank_stock_data.csv')
print("Stock data saved to 'bank_stock_data.csv'.")
```
